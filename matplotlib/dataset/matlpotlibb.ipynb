```python
batsman = pd.read_csv("D:\PYTHON\matplotlib\dataset\sharma-kohli.csv")
print(batsman)
print(type(batsman))
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
```

lets label it , give it a specific title and can perform color opn the graph also title and label 
just use color="you can use it in like red or in hexa decimal "#ffffff")

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="green")
plt.plot(batsman["index"],batsman["RG Sharma"],color="violet")
plt.title("virat kohli and rohit sharma carrier comparison ",color ="red")
plt.xlabel("season")
plt.ylabel("scored")
```

you can alsop use linestyle

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="green",linestyle ="dashed",marker ="o",linewidth =2,markersize =6)
plt.plot(batsman["index"],batsman["RG Sharma"],color="violet",linestyle="dashed",marker ="*", linewidth =2,markersize =5)
plt.title("virat kohli and rohit sharma carrier comparison ",color ="red")
plt.xlabel("season")
plt.ylabel("run_scored")
```

```python
import matplotlib.pyplot as plt
plt.plot(batsman["index"],batsman["V Kohli"],color="green",linestyle ="dashed",marker ="o",linewidth =2,markersize =6,label ="kohli")
plt.plot(batsman["index"],batsman["RG Sharma"],color="violet",linestyle="dashed",marker ="*", linewidth =2,markersize =5,label ="rohit")
plt.title("virat kohli and rohit sharma carrier comparison ",color ="red")
plt.xlabel("season")
plt.ylabel("run_scored")
plt.legend()
#it choose the free space in graph and print it 
#we cne use it like plt.legend(loc ="upper right/ right /left etc ")
plt.grid()#it will add grid line in graph
plt.show()#it will show the graph in screen 
```
